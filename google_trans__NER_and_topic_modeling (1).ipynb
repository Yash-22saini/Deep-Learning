```python
import spacy

# Load English NER model
nlp = spacy.load("en_core_web_lg")
```

```python
from googletrans import Translator
translator = Translator()
```

```python
# Example text
text = "Hello, how are you?"

# Translate to Hindi
result = translator.translate(text, src="en", dest="hi")

print("Original:", result.origin)
print("Translated:", result.text)
print("Source Language:", result.src)
print("Destination Language:", result.dest)
```

```python


texts = [
    "mera naam Yash hai, main jaipur mein rehta hoon , how are you?",
    "Rahul kal Delhi ja raha hai",
    "Modi ji ne Jaipur me speech diya",
    "Priya Amazon company me kaam karti hai",
    "Rohit Sharma ne Mumbai me match jeeta"
]

for i, sentence in enumerate(texts, 1):
    print(f"\n--- Sentence {i} ---")
    print("Original Hinglish:", sentence)

    # Translate Hinglish → English
    translated = translator.translate(sentence, dest="en").text
    print("Translated English:", translated)

    # Apply NER on English text
    doc = nlp(translated)

    print("Named Entities:")
    if doc.ents:
        for ent in doc.ents:
            print(f"{ent.text:<20} | {ent.label_}")
    else:
        print("No entities detected")
```

```python
texts = [
"Cricket is one of the most popular sports in India. Players like Rohit Sharma and Virat Kohli have won many matches for the Indian team. Fans watch cricket matches in stadiums and on television.",
"Technology companies like Amazon, Google, and Microsoft provide many job opportunities. Many engineers work in software development, data science, and artificial intelligence.",
"Food culture in India is very diverse. People enjoy dishes like pizza, burger, pasta, and traditional food such as dal, rice, and roti in restaurants and at home.",
"Politics plays an important role in the development of a country. Leaders give speeches, make policies, and work on economic growth and public welfare."
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# Convert text to numerical form
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(texts)

# Apply Topic Modeling
lda = LatentDirichletAllocation(n_components=4, random_state=42)
lda.fit(X)

# Get words
words = vectorizer.get_feature_names_out()

# Print topics
for i, topic in enumerate(lda.components_):
    print(f"\nTopic {i+1}")
    print([words[j] for j in topic.argsort()[-5:]])
```
