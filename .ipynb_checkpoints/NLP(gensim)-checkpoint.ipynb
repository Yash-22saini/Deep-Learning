```python
import gensim 
import os
```

Gensim : It is an open source library in python written by  Radim Rehurek which is used in unsupervised topic modelling
and natural language processing. It is designed to extract
semantic topics from documents. It can handle large text
collections
.

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('C:\\Users\\user\\Desktop\\game of thrones data'):
    f = open(os.path.join("C:\\Users\\user\\Desktop\\game of thrones data", filename))
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent :
        story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
### story 
model = gensim.models.Word2Vec(
    window = 10 ,
    min_count = 2
)
```

```python
model.build_vocab(story) 
```

```python
print(model.train(story, total_examples = model.corpus_count , epochs = model.epochs)) 
```

```python
 model.wv.most_similar('daenerys')
```

```python
print(model.wv.doesnt_match(['jon' , 'rikon' , 'arya' , 'sansa' , 'bran']) )
```
